```python
df = pd.read_csv("forestfires.csv")
df
```

Changing Nominal Data


#### Change the value of day and month categorical variable

```python
month = ['jan', 'feb', 'mar', 'apr', 'may', 'jun', 'jul', 'aug', 'sep', 'oct', 'nov', 'dec']
day = ['sun', 'mon', 'tue', 'wed', 'thu', 'fri', 'sat']

df['month']   = [month.index(i)+1 for i in df['month']]
df['day']   = [day.index(j)+1 for j in df['day']]
  
df
```

```python
df
```

Getting the number of unique categories 

```python
for att in df.columns:
    if df[att].dtypes == 'object':
        freq = len(df[att].unique())
        print(att, freq)
```

```python
CatagoricalList = ['month', 'day']
for x in CatagoricalList:
        dummies = pd.get_dummies(df[x], prefix=x, dummy_na=False)
        df = df.drop(x, 1)
        df = pd.concat([df, dummies], axis=1)
        
df
```

 Find a strategy to imput missed values


#### Check for missing data

```python
df.isnull().sum().sort_values(ascending=False)
```

#### Impute missing data 


```python
imputer = SimpleImputer(missing_values=ny.nan, strategy='median')
imputer.fit(df)
df = pd.DataFrame(data=imputer.transform(df) , columns=df.columns)
```

Check the missing values again

```python
df.isnull().sum().sort_values(ascending=False)
```

#### Data Visualization

```python
# heatmp =  sb.heatmap(df, cmap="BuGn")
corr = df.corr()
f, ax = plt.subplots(figsize=(20,15))
sb.heatmap(corr, square = True, annot=True, cmap='YlGnBu')
plt.show()
```

```python
# Temprature has a data missing problem and It has been imputed in the previous section.
# Both negative and positive correlation will be selected
# First I found all the correlated values of temprature
# Selected attributes that have a high corrolation with tempreture
cordict = cor.to_dict()
highCor = [i for i in  abs(corr['temp']).to_dict().values() if i > 0.5 and i != 1]
for i in highCor:
    selection1 = list(codict.keys())[list(codict.values()).index(i)]
selection1
```

So far, temprature and RH have been selected. Then, the next step is selecting a cateorical (nominal or ordinal) attribute that
have a great corrolation to tempreture

```python
cor = abs(corr['temp'])
cordict = cor.to_dict()
if cordict['day'] > cordict['month']:
    selection2 = 'day'
else:
     selection2 = 'month'
        
selection2
```
