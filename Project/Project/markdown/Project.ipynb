<h3 style = "color: red"> Natural Language Processing </h3>

```python
# Getting an input file
def getFile (fileName):
    corpus = ''
    with open('../src-data/'+ fileName, encoding='utf8') as f:
        for line in f:
            corpus += ''.join(line.strip())
    return corpus

# Removing punctuations
def rmpunctuation (corpus):
    print('⢿',  end=' ')
    corpus = [char for char in corpus if char in corpus if char not in string.punctuation]
    corpus = ''.join(corpus)
    return corpus

# Removing Numbers
def rmNumber (corpus):
    print('⣻',  end=' ')
    corpus = [char for char in corpus if char in corpus if char not in string.digits]
    corpus = ''.join(corpus)
    return corpus

# Word Tokenization - spliting each word from a list or a string
def wordTokeniz (corpus):
    print('⣽',  end=' ')
    corpus = word_tokenize(corpus)
    return corpus

# Apply stop words
def applyStopWord (corpus):
    print('⣾',  end=' ')
    corpus =  [ word for word in corpus if word.lower() not in stopwords.words('english')]
    return corpus

# Lammetizer -  change a word to its root word
def lammetizer (corpus):
    print('⣷',  end=' ')
    lemma = WordNetLemmatizer()
    corpus = [lemma.lemmatize(word.lower()) for word in corpus]
    return corpus

# Snowball Stemmer - same as Lammetizer - it just chops of suffex of a  word
def stemmer (corpus):
    snow = SnowballStemmer('english')
    corpus = [snow.stem(word.lower()) for word in corpus]
    return corpus
    
# Remove non english words
def onlyASCII (corpus): 
    
    # Remove non Ascii letters
    corpus  = [char for char in corpus if char in corpus if char in string.ascii_letters or char in ' ']
    corpus = ''.join(corpus)
    
#     # Remove non english words
#     CleanCorpus = []
#     English = list(words.words())  #236736 english words
#     for word in corpus:
#         if word in English:
#             CleanCorpus.append(word) 
    return corpus

# Prepars text for each author
def NLP ():
    # File Directory
    files = [file for file in os.listdir('../src-data/') if os.path.isfile(os.path.join('../src-data/',file)) ]
    # Corpuses
    BShelley  = ''
    MShelley  = ''
    JPolidori = ''
    LByron =    ''
    unknown = ''
    # Getting Text data
    for file in files:
        if 'BShelley' in file:
            BShelley += getFile(file)
        if 'MShelley' in file:
            MShelley += getFile(file)
        if 'JPolidori' in file: 
            JPolidori += getFile(file)
        if 'LByron' in file: 
            LByron += getFile(file)
        if 'Frankenstein' in file:
            unknown += getFile(file)
    
    # TEXT PROCESSING
    authors = ['BShelley', 'JPolidori', 'MShelley', 'LByron', 'unknown']
    text = [BShelley, JPolidori, MShelley, LByron, unknown]
    print('Time: ~ 4.5 minutes')
    for i in range(5):
        startTime = time.time()
        print('⣯',  end=' ')
        authorship.loc[authorship['author'] == authors[i], 'text']  =  str(lammetizer(applyStopWord(wordTokeniz(rmNumber(rmpunctuation(text[i]))))))
        print('⣯',  end=' ')
        print (authors[i], ': ', format((time.time() - startTime), '.3f'), 'sec')
    
```

<h3 style = "color: red; font-weight: bold "> Evaluation using Cosine Similarity Between Each Text and Frankenstein </h3>

```python
data
```

```python
def cosSimilarity (data, index):
    corpusList = generateList(data, index)
    unknownList = generateList(data, 4)
    corpus  = ' '.join(corpusList)
    unknown = ' '.join(unknownList)
    document = [corpus, unknown]
    tfidf = TfidfVectorizer().fit_transform(document)
    return cosine_similarity(tfidf[0:1], tfidf)[0,1]
```

### Cosine Similarities with Frankenstein

```python
{'BShelley': cosSimilarity (data, 0), 
 'JPolidori': cosSimilarity (data, 1),
 'MShelley':cosSimilarity (data, 2), 
 'LByron':   cosSimilarity (data, 3), 
 'unknown':  cosSimilarity (data, 4) }
```

The more cosine similarity, the more similarity between the two corpuses. Using cosine similarity, we can see that the value of Marry Wollstonecraft Shelly's word corpus is very higher than other authors' word corpuses. Therefore, the author is Marry Wollstonecraft Shelly.
